```python
import urllib.request, json
import geopandas as gpd
import pandas as pd

import geoplot
import geoplot.crs as gcrs
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(nrows=5, ncols=6, figsize=(30, 30))
fig = plt.figure(figsize=(10,25))

#data france
data_france = gpd.read_file("https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/france.geojson")

last_request = 'https://ptm.huma-num.fr/api/topomine/fantoir/commune/?search=^[a-z\-]{30,50}$|SAINT-REMY-EN-BOUZEMONT&search_method=regex&limit=100000000&offset=0'
last_request_no_tiret = 'https://ptm.huma-num.fr/api/topomine/fantoir/commune/?search=^[a-z]{30,50}$|SAINT-REMY-EN-BOUZEMONT&search_method=regex&limit=100000000&offset=0'

nb_points = []

france_surf = data_france.dissolve()

#loop with Topomine API
for i in range(0,30):

  plot = True

  url_in = 'https://ptm.huma-num.fr/api/topomine/fantoir/commune/?search=^[a-z\-]{'+ str(i+1)+'}$&search_method=regex&limit=100000000&offset=0'
  url_in_no_tiret = 'https://ptm.huma-num.fr/api/topomine/fantoir/commune/?search=^[a-z]{'+ str(i+1)+'}$&search_method=regex&limit=100000000&offset=0'
  # url_in = 'https://ptm.huma-num.fr/api/topomine/fantoir/commune/?search=^[a-z\-]{'+ str(31)+'}$&search_method=regex&limit=100000000&offset=0'

  if(i==29):
    url_in = last_request

  print(url_in)
  with urllib.request.urlopen(url_in) as url:
      response = json.load(url)

  if(response['features'] != None):

    response_str = json.dumps(response)

    # print(response_str)
    gdf = gpd.read_file(response_str, driver='GeoJSON')

  else:
    plot = False
    gdf = gdf.iloc[0:0]

  temp_nb_points = len(gdf)

  col = i % 6
  row = int((i - int(i / 40) * 40) / 6)

  within_points = gpd.sjoin(gdf, france_surf, op = 'within')
  # print(within_points)

  ## traitement (test) no tiret
  with urllib.request.urlopen(url_in_no_tiret) as url:
    response_no_tiret = json.load(url)

  if(response_no_tiret['features'] != None):

    response_no_tiret_str = json.dumps(response_no_tiret)

    # print(response_str)
    gdf_no_tiret = gpd.read_file(response_no_tiret_str, driver='GeoJSON')

  else:
    gdf = gdf.iloc[0:0]

  temp_nb_points_no_tiret = len(gdf_no_tiret)

  nb_points.append((i, temp_nb_points, temp_nb_points_no_tiret))

  # fig, ax = plt.subplots(figsize=(8, 8))
  base = data_france.plot(color='grey', edgecolor='black', linewidth = 0.1, ax=axs[row, col])
  if(plot):
    within_points.plot(ax=base, marker='o', color='red', markersize=0.3);

  axs[row, col].set_xlim(-5, 10)
  axs[row, col].set_ylim(41, 51.5)
  axs[row, col].set_title('n='+str(i+1)+' ('+str(temp_nb_points)+')')
  if(i==29):
    axs[row, col].set_title('n>='+str(i+1)+' ('+str(temp_nb_points)+')')
  axs[row, col].axis('off')

print(nb_points)
```

```python
import matplotlib.pyplot as plt

stat = pd.DataFrame(nb_points, columns=['id', 'Nb communes', 'Nb communes (pas de tiret)'])

# plt.hist(stat, 10)

# stat['nb_communes'].plot(title='Nb de communes en France par nombre de lettres')
# stat['nb_communes_no_tiret'].plot(title='Nb de communes en France par nombre de lettres')

# stat.plot.hist(bins=30, column='nb_communes', title='Nb de communes en France par nombre de lettres')

# Plotting histogram
stat.plot(kind='bar', x='id', y=['Nb communes', 'Nb communes (pas de tiret)'], color=['blue', 'orange'], edgecolor='black', linewidth = 0.1)

plt.xlabel('Nb lettres')
plt.ylabel('Nb communes')
plt.title('Nb de communes en France par nombre de lettres')
```
